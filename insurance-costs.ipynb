```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
#designated filepath
my_filepath = '/kaggle/input/insurance/insurance.csv'
```

```python
#assigning variable to read data
insurance_data = pd.read_csv(my_filepath)
```

```python
#creating scatter plot displaying relationship between BMI and charges w/ smokers singled out
sns.scatterplot(data = insurance_data, x= 'bmi', y='charges', hue='smoker')

#setting name for chart
plt.title("The Relationship Between BMI and Insurance Costs")
```

```python
#chart showing the effect that smoking has on insurance costs despite an individuals age
sns.lmplot(x= 'age', y='charges', hue = 'smoker', data=insurance_data)
```
